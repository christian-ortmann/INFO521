#### Question 1

We will derive the probability that a given observation is part of a bootstrap sample. Suppose that we obtain a bootstrap sample from a set of `n` observations. Don’t forget that samples are obtained with replacement.

  a) What is the probability that the first bootstrap observation is not the $jth$ observation from the original sample? Justify your answer.

> **_Answer:_**  [The probability of bootstrap sample of a specific value ($jth$ value) from sample size $n$ is 1/$n$, thus the probability of NOT selecting a specific value is 1-1/$n$.].

  b) What is the probability that the second bootstrap observation is not the $jth$ observation from the original sample?

> **_Answer:_**  [The probability is 1-1/$n$ for the second observation. This is because we sample with replacement from all $n$ observations, the previous selection has no effect on future selections. Selections are independent.].

  c) Argue that the probability that the $jth$ observation is not in the bootstrap sample is $(1 − 1/n)^n$.

> **_Answer:_**  [Since we know that each bootstrap sample has probability (1-1/$n$) that the selected sample will not be the $jth$ value, we can then multiply that probability for each selection in $n$ samples to get the probability that the $jth$ sample is not in the final bootstrap sample. Mathematically, this is $(1 − 1/n)$ * $(1 − 1/n)$ * $(1 − 1/n)$ ... where each probability is calculated for each single bootstrap sample. If we have 3 samples this is $(1 − 1/3)$ * $(1 − 1/3)$ * $(1 − 1/3)$ which we can then simplify to $(1 − 1/3)^3$ or $(1 − 1/n)^n$ for $n$ samples].

  d) When `n = 5`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  [Since we know that the probability that the sample is NOT in the bootstrap sample is $(1 − 1/n)^n$, we can compute that the probability that it is in the sample is $1-((1 − 1/n)^n)$. In this case, $1-((1 − 1/5)^5)$ which is 0.672].

  e) When `n = 100`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  [Same as 1e. In this case, $1-((1 − 1/100)^100)$ which is 0.634].

  f) When `n = 1000`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  [Same as 1e. In this case, $1-((1 − 1/1000)^1000)$ which is 0.632. As we increase $n$ (limit as n -> inf) our probability becomes asymptotic at 0.632].

#### Question 2

The following questions are relative to k-fold cross-validation.

  a) Explain how k-fold cross-validation is implemented.

> **_Answer:_**  [We break the data into k number folds, or subsets of equal width. We then separate out one of those folds, which we will use as a validation data set. We train a model with the k-1 folds, then use the fold we separated out to test the error of the model fit by the k-1 folds. We do this for each fold, for a total number of k models. We then can average out our error from each validation error from each k-1 model].

  b) How would you choose the analyzed k? Does that matter? (please cite your sources)

> **_Answer:_**  [Choose k so that each fold subset is statistically representative of the overall dataset. We also want to set k so that each fold has the same amount of values. These are done so that model evaluations metrics are computed on datasets of the same size and similar distribution, as well as optimizing computation time. We should also be aware that as k increases, bias decreases but variance will increase.

Source: https://machinelearningmastery.com/k-fold-cross-validation/].

  c) What are the advantages and disadvantages of k-fold cross-validation relative to *the validation set* approach?

> **_Answer:_**  [

    Advantages: 
    - Lower variability, since we are taking the model error multiple times, we may get different errors based on where we split the data.
    - Validation over estimates test error( because we are fitting to only a part of the overall dataset), since with KFolds, we test the entire dataset for test error (an average of all folds)

    Disadvantages: 
    - KFolds is more computationally intensive, which fits k times rather than once for validation
    
].

  d) What are the advantages and disadvantages of k-fold cross-validation relative to *LOOCV*?

> **_Answer:_**  [

    Advantages: 
    - Less computationally intensive since we are only fitting k models rather than n models (LOOCV has k = n)
    - Test error is more accurate because there is less variance among the k models than the n models in LOOCV

    Disadvantages: 
    - LOOCV is better for smaller datasets
    - LOOCV has less bias (advantage) but more variance (disadvantage), due to the number of times we fit the model and how similar each model is from not shaking up the data enough
    - LOOCV is less random than Kfolds (since we fit n times)
    
].

#### Question 3

a) What is the following code supposed to do?

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})

# Number of folds
n_folds = 20

# Polynomial degrees
poly_degrees = range(1, 11)

# KFold cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
mse_estimates = []

# Loop over polynomial degrees
for q in poly_degrees:
    y_hat = np.zeros(n)
    
    for train_idx, test_idx in kf.split(data):
        # Fit on K-1 folds
        poly = PolynomialFeatures(degree=q)
        X_train = poly.fit_transform(data.iloc[train_idx][['x']])
        X_test = poly.transform(data.iloc[test_idx][['x']])
        
        ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
        
        # Predict on the Kth fold
        y_hat[test_idx] = ols.predict(X_test)
    
    # Get the MSE estimate
    mse = np.mean((data['y'] - y_hat) ** 2)
    mse_estimates.append(mse)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(4)
n = 1000
x = np.random.uniform(-4, 4, n)
mu = np.where(x > 0, np.cos(4 * x), 1 - np.sin(x))
y = mu + np.random.normal(size=n)
data = pd.DataFrame({'x': x, 'y': y})

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='black', edgecolor='steelblue', s=50, label='Data', alpha=0.7)
x_ordered = np.sort(x)
mu_ordered = mu[np.argsort(x)]

plt.plot(x_ordered, mu_ordered, color='red', linewidth=3, label="E[Y|X=x], truth")

plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='lower left')
plt.show()
```
