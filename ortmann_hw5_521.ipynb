#### Question 1

We will derive the probability that a given observation is part of a bootstrap sample. Suppose that we obtain a bootstrap sample from a set of `n` observations. Don’t forget that samples are obtained with replacement.

  a) What is the probability that the first bootstrap observation is not the $jth$ observation from the original sample? Justify your answer.

> **_Answer:_**  [The probability of bootstrap sample of a specific value ($jth$ value) from sample size $n$ is 1/$n$, thus the probability of NOT selecting a specific value is 1-1/$n$.].

  b) What is the probability that the second bootstrap observation is not the $jth$ observation from the original sample?

> **_Answer:_**  [The probability is 1-1/$n$ for the second observation. This is because we sample with replacement from all $n$ observations, the previous selection has no effect on future selections. Selections are independent.].

  c) Argue that the probability that the $jth$ observation is not in the bootstrap sample is $(1 − 1/n)^n$.

> **_Answer:_**  [Since we know that each bootstrap sample has probability (1-1/$n$) that the selected sample will not be the $jth$ value, we can then multiply that probability for each selection in $n$ samples to get the probability that the $jth$ sample is not in the final bootstrap sample. Mathematically, this is $(1 − 1/n)$ * $(1 − 1/n)$ * $(1 − 1/n)$ ... where each probability is calculated for each single bootstrap sample. If we have 3 samples this is $(1 − 1/3)$ * $(1 − 1/3)$ * $(1 − 1/3)$ which we can then simplify to $(1 − 1/3)^3$ or $(1 − 1/n)^n$ for $n$ samples].

  d) When `n = 5`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  [Since we know that the probability that the sample is NOT in the bootstrap sample is $(1 − 1/n)^n$, we can compute that the probability that it is in the sample is $1-((1 − 1/n)^n)$. In this case, $1-((1 − 1/5)^5)$ which is 0.672].

  e) When `n = 100`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  [Same as 1e. In this case, $1-((1 − 1/100)^100)$ which is 0.634].

  f) When `n = 1000`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  [Same as 1e. In this case, $1-((1 − 1/1000)^1000)$ which is 0.632. As we increase $n$ (limit as n -> inf) our probability becomes asymptotic at 0.632].

#### Question 2

The following questions are relative to k-fold cross-validation.

  a) Explain how k-fold cross-validation is implemented.

> **_Answer:_**  [We break the data into k number folds, or subsets of equal width. We then separate out one of those folds, which we will use as a validation data set. We train a model with the k-1 folds, then use the fold we separated out to test the error of the model fit by the k-1 folds. We do this for each fold, for a total number of k models. We then can average out our error from each validation error from each k-1 model].

  b) How would you choose the analyzed k? Does that matter? (please cite your sources)

> **_Answer:_**  [Choose k so that each fold subset is statistically representative of the overall dataset. We also want to set k so that each fold has the same amount of values. These are done so that model evaluations metrics are computed on datasets of the same size and similar distribution, as well as optimizing computation time. We should also be aware that as k increases, bias decreases but variance will increase.

Source: https://machinelearningmastery.com/k-fold-cross-validation/].

  c) What are the advantages and disadvantages of k-fold cross-validation relative to *the validation set* approach?

> **_Answer:_**  [

    Advantages: 
    - Lower variability, since we are taking the model error multiple times, we may get different errors based on where we split the data.
    - Validation over estimates test error( because we are fitting to only a part of the overall dataset), since with KFolds, we test the entire dataset for test error (an average of all folds)

    Disadvantages: 
    - KFolds is more computationally intensive, which fits k times rather than once for validation
    
].

  d) What are the advantages and disadvantages of k-fold cross-validation relative to *LOOCV*?

> **_Answer:_**  [

    Advantages: 
    - Less computationally intensive since we are only fitting k models rather than n models (LOOCV has k = n)
    - Test error is more accurate because there is less variance among the k models than the n models in LOOCV

    Disadvantages: 
    - LOOCV is better for smaller datasets
    - LOOCV has less bias (advantage) but more variance (disadvantage), due to the number of times we fit the model and how similar each model is from not shaking up the data enough
    - LOOCV is less random than Kfolds (since we fit n times)
    
].

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})

# Number of folds
n_folds = 20

# Polynomial degrees
poly_degrees = range(1, 11)

# KFold cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
mse_estimates = []

# Loop over polynomial degrees
for q in poly_degrees:
    y_hat = np.zeros(n)
    
    for train_idx, test_idx in kf.split(data):
        # Fit on K-1 folds
        poly = PolynomialFeatures(degree=q)
        X_train = poly.fit_transform(data.iloc[train_idx][['x']])
        X_test = poly.transform(data.iloc[test_idx][['x']])
        
        ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
        
        # Predict on the Kth fold
        y_hat[test_idx] = ols.predict(X_test)
    
    # Get the MSE estimate
    mse = np.mean((data['y'] - y_hat) ** 2)
    mse_estimates.append(mse)
```

```python
# BEGIN SOLUTION
#changes denoted with full caps comments

#-----------------------------------------------------------------------------------------
#IMPORTS
#-----------------------------------------------------------------------------------------
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error


#-----------------------------------------------------------------------------------------
#DATA SIMULATION
#-----------------------------------------------------------------------------------------
np.random.seed(0) #set a random seed to generate data 
n = 100 #size of data set
order = 2 #polynomial degree to apply to y, relationship between x and y
x = np.random.uniform(-4, 4, n) #create x, starting from -4 and ending at 4, with 100 values
y = np.random.normal(loc=x**order, scale=0.25, size=n) #create y, creating a dataset that has median at x^2, SD of 0.25, and size of 100 (n)
data = pd.DataFrame({'x': x, 'y': y})

# Polynomial degrees
#CREATE MORE CLEAR VARIABLES TO SET RANGE
start = 1 #sequence of integers from 1-10, here start at 1
stop = 11 #sequence of integers from 1-10, here end at 11 because we want 1-10, but python indexes from 0
poly_degrees = range(start, stop) #sequence of integers from 1-10

#-----------------------------------------------------------------------------------------
#MODEL FITTING WITH K-FOLD CV
#-----------------------------------------------------------------------------------------
# Number of folds for k-fold cross validation
n_folds = 20
# KFold cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
mse_estimates = [] #SET AN EMPTY LIST TO STORE OUR MSE ESTIMATES AS WE CREATE THEM

# Loop over polynomial degrees
for q in poly_degrees:
    y_hat = np.zeros(n)
    
    for train_idx, test_idx in kf.split(data):
        # Fit on K-1 folds
        poly = PolynomialFeatures(degree=q)
        X_train = poly.fit_transform(data.iloc[train_idx][['x']])
        X_test = poly.transform(data.iloc[test_idx][['x']])
        
        ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
        
        # Predict on the Kth fold
        y_hat[test_idx] = ols.predict(X_test)
    
    # Get the MSE estimate
    #mse = np.mean((data['y'] - y_hat) ** 2)
    mse = mean_squared_error(data['y'],y_hat) #USED THE MSE FUNCTION FROM SCIKIT-LEARN
    mse_estimates.append(mse)


#-----------------------------------------------------------------------------------------
#PLOTS
#-----------------------------------------------------------------------------------------


#PLOT MSE vs POLY DEGREE
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax1 = plt.subplots()
sns.lineplot(x = poly_degrees, y = mse_estimates, ax=ax1,color = 'red')
plt.xticks(poly_degrees); #to show all poly degrees
plt.xlabel('Poly Degree')
ax1.set_ylabel('MSE', color = 'red')
plt.title('MSE vs Polynomial Degree')

ax2 = ax1.twinx() #plot a second y axis
sns.lineplot(x =list(poly_degrees)[1:], y = mse_estimates[1:], ax = ax2, color = 'blue') #remove polynomial = 1 to show change in more detail
ax2.set_ylabel('MSE, degree = 1 omitted', color = 'blue');

#PLOT X VS Y

plt.figure()
sns.lineplot(x = x, y = y, label = 'Original Data')

#rerun the model and just show the overall fit by predicting on the training data
X = poly.fit_transform(data[['x']])
ols = sm.OLS(data['y'], X).fit()
y_hat = ols.predict(X) #predict on training data to show model fit (test error should not be interpreted from this prediction)
sns.lineplot(x = x, y = y_hat, label = ' Poly Degree = 2\n*Model Fit Shown*')
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y');


# END SOLUTION
```

#### Question 4

Bootstrap the following dataset (`n = 1000`) to obtain median and 95% CI for parameter estimates (slope and intercept) summarizing the relationship between `x` and `y_measured.` What happens with median parameter estimates when you examine `y` instead? Provide some visualizations to explain your results.

```python
import numpy as np
import statsmodels.api as sm
import scipy.stats as st


np.random.seed(1)
nobs = 1000
x = np.random.normal(size=nobs)
y = 5 - 2.25 * x
y_measured = y + np.random.normal(size=nobs)

# BEGIN SOLUTION

nboots = 1000

x = pd.DataFrame(x, columns = ['x'])
y = pd.DataFrame(y, columns = ['y'])
y_measured = pd.DataFrame(y_measured, columns = ['y_measured'])

#for testing y_measured
intMeas = []
slopeMeas = []

for _ in range(nboots):
    #bootstrap x source for sample method https://medium.com/@whystudying/resampling-with-python-bootstrap-50f21866d7c9
    xBoot = x.sample(n = 1000, replace = True)
    #bootstrap y
    yBoot = y.sample(n = 1000, replace = True)
     #bootstrap y_measured
    yMeasBoot = y_measured.sample(n = 1000, replace = True)
    #reset indicies and combine them into one Dataframe
    bootData = pd.concat([xBoot.reset_index(drop=True), yMeasBoot.reset_index(drop=True), y.reset_index(drop=True)], axis=1)
    X = sm.add_constant(bootData[['x']])
    ols = sm.OLS(bootData['y_measured'], X).fit()
    # Append results
    intMeas.append(ols.params['const'])
    slopeMeas.append(ols.params['x'])

#source: https://www.geeksforgeeks.org/how-to-calculate-confidence-intervals-in-python/
print(f'X and Y_Measured Bootstrapped 1000 times\n95% Confidence Intervals: \n  Intercept:{st.norm.interval(confidence = 0.95, loc=np.mean(intMeas), scale=st.sem(intMeas))}\n  Slope:{st.norm.interval(confidence = 0.95, loc=np.mean(slopeMeas), scale=st.sem(slopeMeas))} \nMedian: \n  Intercept:{np.median(intMeas)}\n  Slope:{np.median(slopeMeas)}')

#for testing y
int = []
slope = []


for _ in range(nboots):
    #bootstrap x source for sample method https://medium.com/@whystudying/resampling-with-python-bootstrap-50f21866d7c9
    xBoot = x.sample(n = 1000, replace = True)
    #bootstrap y
    yBoot = y.sample(n = 1000, replace = True)
     #bootstrap y_measured
    yMeasBoot = y_measured.sample(n = 1000, replace = True)
    #reset indicies and combine them into one Dataframe
    bootData = pd.concat([xBoot.reset_index(drop=True), yMeasBoot.reset_index(drop=True), yBoot.reset_index(drop=True)], axis=1)
    X = sm.add_constant(bootData[['x']])
    ols = sm.OLS(bootData['y'], X).fit()
    # Append results
    int.append(ols.params['const'])
    slope.append(ols.params['x'])

print(f'\nX and Y Bootstrapped 1000 times\n95% Confidence Intervals: \n  Intercept:{st.norm.interval(confidence = 0.95, loc=np.mean(int), scale=st.sem(int))}\n  Slope:{st.norm.interval(confidence = 0.95, loc=np.mean(slope), scale=st.sem(slope))} \nMedian: \n  Intercept:{np.median(int)}\n  Slope:{np.median(slope)}')

#-----------------------------------------------------------------------------------------
#PLOTS
#-----------------------------------------------------------------------------------------

plt.figure()
sns.histplot(slope, color="orange", label="x vs y", kde=True)
sns.histplot(slopeMeas, color="blue", label="x vs y_measured", kde=True)
plt.title('Distribution of Slope Estimates')
plt.xlabel('Slope')
plt.legend()


plt.figure()
sns.histplot(int, color="orange", label="x vs y", kde=True)
sns.histplot(intMeas, color="blue", label="x vs y_measured", kde=True)
plt.title('Distribution of Intercept Estimates')
plt.xlabel('Intercept')
plt.legend();


# END SOLUTION
```
